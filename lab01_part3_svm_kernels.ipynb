## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

Describe results in one-two sentences.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

# Ð“ÐµÐ½ÐµÑ€Ð°Ñ†Ð¸Ñ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ð° Moons
from sklearn.datasets import make_moons
moons_points, moons_labels = make_moons(n_samples=500, noise=0.2, random_state=42)

# Ð›Ð¾Ð³Ð¸ÑÑ‚Ð¸Ñ‡ÐµÑÐºÐ°Ñ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ñ Ñ Ð³Ð¸Ð¿ÐµÑ€Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð°Ð¼Ð¸
lr = LogisticRegression(C=1.0, solver='lbfgs', max_iter=300, random_state=42)
lr.fit(moons_points, moons_labels)
print("Accuracy (Logistic Regression):", accuracy_score(moons_labels, lr.predict(moons_points)))

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¾Ð±Ð»Ð°ÑÑ‚ÐµÐ¹ Ñ€ÐµÑˆÐµÐ½Ð¸Ð¹ Ð´Ð»Ñ Ð»Ð¾Ð³Ð¸ÑÑ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ð¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸
plt.figure(figsize=(9, 5))
plot_decision_regions(moons_points, moons_labels, clf=lr, legend=2, colors="lightcoral,lightblue")
plt.title("Logistic Regression Decision Boundary", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.grid(alpha=0.3, linestyle=':')
plt.show()

# SVM Ñ Ð»Ð¸Ð½ÐµÐ¹Ð½Ñ‹Ð¼ ÑÐ´Ñ€Ð¾Ð¼ Ð¸ Ð³Ð¸Ð¿ÐµÑ€Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð°Ð¼Ð¸
svm_linear = SVC(kernel='linear', C=0.5, random_state=42)
svm_linear.fit(moons_points, moons_labels)
print("Accuracy (Linear SVM):", accuracy_score(moons_labels, svm_linear.predict(moons_points)))

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¾Ð±Ð»Ð°ÑÑ‚ÐµÐ¹ Ñ€ÐµÑˆÐµÐ½Ð¸Ð¹ Ð´Ð»Ñ SVM
plt.figure(figsize=(9, 5))
plot_decision_regions(moons_points, moons_labels, clf=svm_linear, legend=2, colors="lightgreen,lightyellow")
plt.title("Linear SVM Decision Boundary", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.grid(alpha=0.3, linestyle=':')
plt.show()
```

## 1.2 Kernel tirck

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

# SVM Ñ Ð¿Ð¾Ð»Ð¸Ð½Ð¾Ð¼Ð¸Ð°Ð»ÑŒÐ½Ñ‹Ð¼ ÑÐ´Ñ€Ð¾Ð¼
svm_poly = SVC(kernel='poly', degree=3, C=1, random_state=42)
svm_poly.fit(moons_points, moons_labels)
print("Accuracy (Polynomial Kernel SVM):", accuracy_score(moons_labels, svm_poly.predict(moons_points)))

plt.figure(figsize=(10, 5))
plot_decision_regions(moons_points, moons_labels, clf=svm_poly, legend=2, colors="plum,skyblue")
plt.title("Polynomial Kernel SVM Decision Boundary", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.grid(alpha=0.3, linestyle=':')
plt.show()

# SVM Ñ RBF ÑÐ´Ñ€Ð¾Ð¼
svm_rbf = SVC(kernel='rbf', C=1, random_state=42)
svm_rbf.fit(moons_points, moons_labels)
print("Accuracy (RBF Kernel SVM):", accuracy_score(moons_labels, svm_rbf.predict(moons_points)))

plt.figure(figsize=(10, 5))
plot_decision_regions(moons_points, moons_labels, clf=svm_rbf, legend=2, colors="lavender,lightgreen")
plt.title("RBF Kernel SVM Decision Boundary", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.grid(alpha=0.3, linestyle=':')
plt.show()

# SVM Ñ ÑÐ¸Ð³Ð¼Ð¾Ð¸Ð´Ð½Ñ‹Ð¼ ÑÐ´Ñ€Ð¾Ð¼
svm_sigmoid = SVC(kernel='sigmoid', C=1, random_state=42)
svm_sigmoid.fit(moons_points, moons_labels)
print("Accuracy (Sigmoid Kernel SVM):", accuracy_score(moons_labels, svm_sigmoid.predict(moons_points)))

plt.figure(figsize=(10, 5))
plot_decision_regions(moons_points, moons_labels, clf=svm_sigmoid, legend=2, colors="lightyellow,lightcoral")
plt.title("Sigmoid Kernel SVM Decision Boundary", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.grid(alpha=0.3, linestyle=':')
plt.show()
```

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

# Ð¡Ð¾Ð·Ð´Ð°ÐµÐ¼ Pipeline Ð´Ð»Ñ Ð›Ð¾Ð³Ð¸ÑÑ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ð¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸ Ñ Ð¿Ð¾Ð»Ð¸Ð½Ð¾Ð¼Ð¸Ð°Ð»ÑŒÐ½Ñ‹Ð¼Ð¸ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ°Ð¼Ð¸
poly_log_reg = Pipeline([
    ("poly_features", PolynomialFeatures(degree=4, include_bias=False)),  # Ð£Ð²ÐµÐ»Ð¸Ñ‡Ð¸Ð»Ð¸ ÑÑ‚ÐµÐ¿ÐµÐ½ÑŒ Ð´Ð¾ 4, ÑƒÐ±Ñ€Ð°Ð»Ð¸ bias
    ("log_reg", LogisticRegression(C=1.0, solver='lbfgs', max_iter=300, random_state=42))  # Ð”Ð¾Ð±Ð°Ð²Ð¸Ð»Ð¸ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹
])

# ÐžÐ±ÑƒÑ‡ÐµÐ½Ð¸Ðµ Ð¼Ð¾Ð´ÐµÐ»Ð¸
poly_log_reg.fit(moons_points, moons_labels)

# ÐžÑ†ÐµÐ½ÐºÐ° Ñ‚Ð¾Ñ‡Ð½Ð¾ÑÑ‚Ð¸ Ð¼Ð¾Ð´ÐµÐ»Ð¸
accuracy = accuracy_score(moons_labels, poly_log_reg.predict(moons_points))
print(f"Accuracy (Polynomial Logistic Regression): {accuracy:.3f}")

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¾Ð±Ð»Ð°ÑÑ‚ÐµÐ¹ Ñ€ÐµÑˆÐµÐ½Ð¸Ð¹
plt.figure(figsize=(10, 6))  # Ð£Ð²ÐµÐ»Ð¸Ñ‡Ð¸Ð»Ð¸ Ñ€Ð°Ð·Ð¼ÐµÑ€ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ°
plot_decision_regions(moons_points, moons_labels, clf=poly_log_reg, legend=2, colors="lightpink,lightcyan")
plt.title("Polynomial Logistic Regression Decision Boundary", fontsize=14, fontweight='bold')
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.grid(alpha=0.3, linestyle=':')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions

# ÐžÐ±ÑŠÐµÐ´Ð¸Ð½ÐµÐ½Ð¸Ðµ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ð¾Ð²
points = np.vstack((circles_points * 2.2 + 0.7, moons_points))
labels = np.hstack((circles_labels, moons_labels + 2))  # +2 Ð´Ð»Ñ Ñ€Ð°Ð·Ð´ÐµÐ»ÐµÐ½Ð¸Ñ ÐºÐ»Ð°ÑÑÐ¾Ð²

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¾Ð±ÑŠÐµÐ´Ð¸Ð½Ñ‘Ð½Ð½Ð¾Ð³Ð¾ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ð°
plt.figure(figsize=(4, 4))  # Ð£Ð¼ÐµÐ½ÑŒÑˆÐµÐ½Ð½Ñ‹Ð¹ Ñ€Ð°Ð·Ð¼ÐµÑ€ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ°
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap='coolwarm', edgecolor='black', s=20)
plt.title("Combined Circles and Moons Dataset", fontsize=12, fontweight='bold')
plt.xlabel("Feature 1", fontsize=10)
plt.ylabel("Feature 2", fontsize=10)
plt.grid(alpha=0.3, linestyle=':')
plt.tight_layout()
plt.show()

# ÐžÐ±ÑƒÑ‡ÐµÐ½Ð¸Ðµ Ð¼Ð¾Ð´ÐµÐ»Ð¸ SVM Ñ RBF ÑÐ´Ñ€Ð¾Ð¼
svm_rbf_combined = SVC(kernel='rbf', C=1, random_state=42)
svm_rbf_combined.fit(points, labels)
print("Accuracy (RBF Kernel SVM) on Combined Dataset:", accuracy_score(labels, svm_rbf_combined.predict(points)))

# Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð¾Ð±Ð»Ð°ÑÑ‚ÐµÐ¹ Ñ€ÐµÑˆÐµÐ½Ð¸Ð¹
plt.figure(figsize=(6, 5))  # Ð£Ð¼ÐµÑ€ÐµÐ½Ð½Ñ‹Ð¹ Ñ€Ð°Ð·Ð¼ÐµÑ€ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ°
plot_decision_regions(points, labels, clf=svm_rbf_combined, legend=2, colors="pink,lightblue,lightgreen")
plt.title("RBF Kernel SVM Decision Boundary", fontsize=12, fontweight='bold')
plt.xlabel("Feature 1", fontsize=10)
plt.ylabel("Feature 2", fontsize=10)
plt.grid(alpha=0.3, linestyle=':')
plt.tight_layout()
plt.show()
```
